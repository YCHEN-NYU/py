# Energy Barriers from Dmitry's PRB 76, 054410

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

fig = plt.figure(figsize=(8,6))
fig.patch.set_facecolor('white')
h = np.linspace(0,0.15,1000)
u = (1-h)**2 - 0.0806*(1-abs(np.sin(np.pi*h*20)))**2
plt.xlabel('h')
plt.ylabel('u(h)')
plt.plot(h, u, 'b.')
plt.grid(True)
plt.show()
```

# Energy Barrier without considering quantum tunneling resonances

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
fontAX = 14
fontTitle = 20
fig = plt.figure(figsize=(8,6))
fig.patch.set_facecolor('white')
sz = np.linspace(-1,1,100)
D = 0.708
S = 10
E = -(sz**2+sz)*D*S**2
plt.xlabel('s_z', fontsize=fontAX)
plt.ylabel('$E(s_z)$', fontsize=fontAX)
plt.title(' $\quad E(s_z) - s_z$', fontsize=fontTitle)
plt.plot(sz, E, 'r-')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
fontAX = 14
fontTitle = 20
fig = plt.figure(figsize=(8,6))
fig.patch.set_facecolor('white')
ax = plt.subplot(111)

hz = np.linspace(0,0.1,1000)
hx = np.linspace(0.2,0.4,1000)
u = (1-hz)**2 - (1-hz)*hx
# plt.plot(hz, u, 'b', linewidth=2)
plt.plot(hx, u, 'r', linewidth=2)
plt.title('$u(h) - h$', fontsize=fontTitle)
plt.xlabel('h', fontsize=fontAX)
plt.ylabel('u(h)', fontsize=fontAX)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy.random as rnd
from matplotlib.patches import Ellipse
# %matplotlib inline

NUM = 500

ells = [Ellipse(xy=rnd.rand(2)*30, width=rnd.rand(),
                height=rnd.rand(), angle=rnd.rand()*360)
        for i in range(NUM)]

fig = plt.figure( figsize=(10.8, 19.2), facecolor='w', )
ax = fig.add_subplot(111, aspect='equal')
for e in ells:
    ax.add_artist(e)
    e.set_clip_box(ax.bbox)
    e.set_alpha(rnd.rand())
    e.set_facecolor(rnd.rand(3))

ax.set_xlim(0, 10.8)
ax.set_ylim(0, 19.2)
plt.axis('off')
plt.show()
fig.savefig('foo.png')
```
